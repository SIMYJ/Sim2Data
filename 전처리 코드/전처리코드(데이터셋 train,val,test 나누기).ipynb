````python
import os
from glob import glob
import shutil
from sklearn.model_selection import train_test_split
import os

image_files = glob("./ME/train_label/*.jpg")
#print(image_files[0])
print(len(image_files))

image = [name.replace(".jpg","") for name in image_files] 
#print(image[0])
train_names, test_names = train_test_split(image, test_size= 0.2, random_state= 42, shuffle=True)

val_names, test_names = train_test_split(test_names, test_size= 0.5, random_state= 42, shuffle=True)

print(len(train_names))
print(len(test_names))
print(len(val_names))

def batch_move_files(file_list, source_path, destination_path):
    for file in file_list:
        image = file.split('/')[-1] + '.jpg'
        print(image)
        txt = file.split('/')[-1] + '.txt'
        print(source_path)
        print(os.path.join(source_path, image))

        shutil.copy(os.path.join(source_path, image), os.path.join(destination_path, image))
        shutil.copy(os.path.join(source_path, txt), os.path.join(destination_path, txt))
    return

# data path
source_dir ="./ME/train_label/"

# new data path
train_dir = './tmp_repo/train/'
test_dir = './tmp_repo/test/'
val_dir = './tmp_repo/val/'


batch_move_files(train_names, source_dir, train_dir)

batch_move_files(test_names, source_dir, test_dir)

batch_move_files(val_names, source_dir, val_dir)

print("done")
````

## 2.2) 욜로규격 경로 설정

파일구조
```
상위폴더
|---전처리코드.ipynb
|
|---Me
|    
|    |---train_label         *.jpg,*.txt(trian이미지와 라벨이미지 존재)
|
|---data
|                           (아래 코드를 실행 시키면 train : val : test = 8 :1 :1 로 데이터셋 나눔 
|    |---images               
|         |
|         |---train         *.jpg
|         |---vaild         *.jpg
|         |---vaild         *.jpg
|    
|    |---label                 
|         |
|         |---train         *.txt
|         |---vaild         *.txt
|         |---vaild         *.txt
|
|__________

```


````python
import os
from glob import glob
import shutil
from sklearn.model_selection import train_test_split
import os

image_files = glob("./ME/train_label/*.jpg")
#print(image_files[0])
print('총 데이터량: {}'.format(len(image_files)))

image = [name.replace(".jpg","") for name in image_files] 
#print(image[0])
train_names, test_names = train_test_split(image, test_size= 0.2, random_state= 42, shuffle=True)

val_names, test_names = train_test_split(test_names, test_size= 0.5, random_state= 42, shuffle=True)

print("train set: {}".format(len(train_names)))
print("test set: {}".format(len(test_names)))
print("val set: {}".format(len(val_names)))

dataset_names = [train_names,val_names,test_names]
dataset_kinds = ["train/","valid/","test/"]

def Batch_move_files_yolo_img(dataset_names,dataset_kinds, source_path, destination_path):
    for img_list, kind in zip(dataset_names, dataset_kinds):
        for img in img_list:
            image = img.split('/')[-1] + '.jpg'
            
            shutil.copy(os.path.join(source_path, image), os.path.join(destination_path + kind, image))
    return

def Batch_move_files_yololbel(dataset_names,dataset_kinds, source_path, destination_path):
    for txt_list, kind in zip(dataset_names, dataset_kinds):
        for txt in txt_list:
            #print(file)
            txt = txt.split('/')[-1] + '.txt'
            
            shutil.copy(os.path.join(source_path, txt), os.path.join(destination_path + kind, txt))
    return


# data path
source_dir ="./ME/train_label/"

# new data path
img_dir = './data/images/'
label_dir = './data/labels/'


Batch_move_files_yolo_img(dataset_names, dataset_kinds , source_dir, img_dir)

Batch_move_files_yololbel(dataset_names, dataset_kinds , source_dir, label_dir)



print("done")
````
